```python
import cupy as cp
```

```python
import numpy as np
```

```python
array_cpu = np.random.randint(0, 255, size=(200, 200))
```

```python
array_cpu.nbytes / 1e6
```

```python
array_gpu = cp.asarray(array_cpu)
```

```python
%%timeit
cp.asarray(array_cpu)
```

```python
type(array_gpu)
```

```python
from scipy import fft
```

```python
%%timeit
fft.fftn(array_cpu)
```

```python
from cupyx.scipy import fft as fft_gpu
```

```python
%%timeit
fft_gpu.fftn(array_gpu)
```

```python
%%timeit
array_cpu_again = cp.asnumpy(fft_gpu.fftn(array_gpu))
```

```python
array_cpu_again = cp.asnumpy(fft_gpu.fftn(array_gpu))
```

```python
np.allclose(array_cpu_again, fft.fftn(array_cpu))
```
