```python
from numba import cuda
import cupy as cp 
import math as m 

@cuda.jit
def my_kernel2D(io_array):
    x, y = cuda.grid(2)
    if x < io_array.shape[0] and y < io_array.shape[1]:
        io_array[x, y] *= 2

data = cp.ones(shape=(16, 16))
threadsperblock = (16, 16)
blockspergrid_x = m.ceil(data.shape[0] / threadsperblock[0])
blockspergrid_y = m.ceil(data.shape[1] / threadsperblock[1])
blockspergrid = (blockspergrid_x, blockspergrid_y)

my_kernel2D[blockspergrid, threadsperblock](data)

print(data)
```
