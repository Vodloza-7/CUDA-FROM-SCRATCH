```python
from numba import jit   # numba module used to compile CPU functions
import math as m 

@jit
def hypot(x, y):
    # Implementation of hypot can be found here https://en.wikipedia.org/wiki/Hypot
    x, y = abs(x), abs(y)
    t, x = min(x, y), max(x, y)
    t /= x
    return x * m.sqrt(1 + t * t)
```

```python
hypot(3.0, 4.0)
```

```python
hypot.py_func(3.0, 4.0)
```

```python
%%timeit
hypot(3.0, 4.0)     # 245 ns +- 4.43 ns per loop
```

```python
%%timeit
hypot.py_func(3.0, 4.0)     # 565 ns +- 7.16 ns per loop
```

```python
%%timeit
m.hypot(3.0, 4.0)       # 118 ns +- 24.9 ns per loop
```
