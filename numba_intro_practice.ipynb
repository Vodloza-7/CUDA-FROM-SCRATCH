```python
from numba import cuda
import cupy as cp
```

```python
@cuda.jit
def addOneKernel(A):
    i = cuda.grid(1)
    if i < A.shape[0]:
        A[i] += 1
```

```python
my_array = cp.ones(256)
threadPerBlock = 32
```

```python
blocksPerGrid = (my_array.size * (threadPerBlock - 1)) // threadPerBlock
# blocksPerGrid =  my_array.size / threadPerBlock
```

```python
execution_threads = threadPerBlock * blocksPerGrid
```

```python
addOneKernel[threadPerBlock, blocksPerGrid](my_array)
```
